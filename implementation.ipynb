```python
filepath = r"C:/Users/ak52227/Downloads/"
my_project_name = r"my_transmon_2"
my_design_name = "transmon_no_pins_2"
my_setup_names = ["EigenmodeSetup_2"]
```

```python
# ---------------------------------------------------------
# 1. Conenct to Ansys HFSS and generate your geometry
# --------------------------------------------------------- 

Desktop(specified_version="2022.2", new_desktop=False)

my_hfss = ep.hfss_reinhold(
    project_name=filepath+my_project_name+".aedt",
    design_name=my_design_name,
    setup_names=my_setup_names,)

my_hfss.start_hfss_file()

my_hfss.create_hfss_geometries()
my_hfss.hfss.save_project()
```

```python
# ---------------------------------------------------------
# 2. Generate mesh and run eigenmode simulation.
# --------------------------------------------------------- 

my_hfss.hfss.mesh.generate_mesh(my_setup_names[0])
my_hfss.hfss.analyze_setup(my_setup_names[0])
my_hfss.hfss.save_project()
```

```python
# ---------------------------------------------------------
# 3. Start pyepr distributed analysis
# ---------------------------------------------------------
tmon = ep.pyepr_transmon()

# These depend on your design!!
tmon.snail_mode = 0
tmon.cavity_mode = 1
tmon.box_mode = 2
tmon.readout_mode = 3

tmon.distributed_analysis(filepath, my_project_name, my_design_name)

# The distributed analysis produces a file:
distributed_file = tmon.distributed_file
```

```python
# ---------------------------------------------------------
# 4. Run quantum analysis
# ---------------------------------------------------------

tmon.quantum_analysis(distributed_file, new_folder_path=filepath)
```

```python
# ---------------------------------------------------------
# 5. Calculate mixing (3-wave or 4-wave) if you are workign with a SNAIL. SKIP IF YOU ARE USING A SIMPLE TRANSMON!
# ---------------------------------------------------------

# Parameters
n = 3            # SNAIL number of small junctions
M = 1            # number of unit cells
a = 0.219        # asymmetry parameter
Phi_ratio = 0.5  # external flux / Phi0
Lj = 1.8e-9      # junction inductance (H)
var = 0          # variation index

# ---- 4-WAVE MIXING ----
K4, Kerr_r, Kerr_c, Kerr_box, g3_4, g4_4, Ls, L0 = tmon.main_4wave_mixing(
    n, M, a, Phi_ratio, Lj, var
)

print("4-wave Kerr =", K4)
print("g4 =", g4_4)

# ---- 3-WAVE MIXING ----
K3, Kerr_sc, Kerr_box3, Kerr_r3, g3_3, g4_3 = tmon.main_3wave_mixing(
    n, M, a, Phi_ratio, Lj, var
)

print("3-wave Kerr =", K3)
print("g3 =", g3_3)
```

```python
# ---------------------------------------------------------
# 6. Saving all Kerr and mixing data for SNAILs
# ---------------------------------------------------------

snail_data_4, snail_data_3= tmon.save_quantum_results(new_folder_path=filepath)

print("CSV files saved in:", filepath)
```
